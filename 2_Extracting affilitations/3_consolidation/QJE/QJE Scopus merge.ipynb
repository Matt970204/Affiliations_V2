```python
#Read in the master sheet
#Local store of AER_master
QJE_master = pd.read_excel('/Users/MattJohnson/Desktop/Affiliations/1_Input files and processing input files/1_1_Input files/Master lists/Processed/QJE_processed.xlsx')
for i in range(len(QJE_master)):
    QJE_master.loc[i,['Article_ID']] = QJE_master.loc[i,['stable_url']].to_string().rsplit('/',1)[-1]

QJE_master_condenced  = QJE_master[['Article_ID','year','authors','content_type']]
QJE_master_condenced['year'] = QJE_master_condenced['year'].astype('int')
QJE_master_condenced['Article_ID'] = QJE_master_condenced['Article_ID'].astype('int')
QJE_master_condenced.rename(columns = {'year':'year_master', 'content_type':'content_type_master'}, inplace = True)
mask_year = QJE_master_condenced['year_master'].isin([2016,2015,2014,2013,2012,2011])
QJE_master_condenced = QJE_master_condenced[~mask_year]
#mask_type = QJE_master_condenced['content_type_master'].isin(['Rejoiner','MISC','Review'])
#QJE_master_condenced = QJE_master_condenced[~mask_type]
mask_type = QJE_master_condenced['content_type_master'].isin(['Article',])
QJE_master_condenced = QJE_master_condenced[mask_type]
#QJE_master_condenced.drop(~QJE_master_condenced.loc[QJE_master_condenced['year_master'].isin([2016])].index, inplace=True)
print(QJE_master_condenced.shape)
QJE_master_condenced.head()
```

```python
# Merging the masterfile onto the new melted file to get additional data
consolidated_master_merged = QJE_master_condenced.merge(QJE_consolidated, how = 'left', on = 'Article_ID')
consolidated_master_merged = consolidated_master_merged.sort_values(by=['year_master'])
consolidated_master_merged = consolidated_master_merged.reset_index(drop = True)
```
