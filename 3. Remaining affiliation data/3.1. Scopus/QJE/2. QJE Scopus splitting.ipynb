```python
import pandas as pd
QJE_scopus = pd.read_excel('/Users/MattJohnson/Desktop/Affiliations/1_Input files and processing input files/1_1_Input files/Scopus files/QJE_SCOPUS.xlsx')
print(QJE_scopus.shape)
QJE_scopus = QJE_scopus.loc[QJE_scopus['year'] >= 1980]
QJE_scopus = QJE_scopus.sort_values(by=['year'])
QJE_scopus = QJE_scopus.reset_index(drop=True)
print(QJE_scopus.shape)
```

```python
import re
from rapidfuzz import fuzz
import numpy as np
#for i in range(1577,1578):
QJE_scopus = QJE_scopus.replace(np.nan, '', regex=True)

association_list = ['a--','b--','c--','d--','e--','f--','g--','h--']
#for i in range(1577,1578):
#for i in range(1594,1595):
#for i in range(2846,2847):
for i in range(len(QJE_scopus)):
    print(i)

    #print(QJE_scopus.loc[i,'authors'])
    #print(QJE_scopus.loc[i,'affiliations'])

# Checking for if the authors and affiliations are split by a-- b-- c--
    ratio = fuzz.partial_ratio(QJE_scopus.loc[i,'authors'] , 'a--')
    if ratio < 95:
        QJE_scopus.loc[i,'0_author'] = QJE_scopus.iloc[i]['authors']
        QJE_scopus.loc[i,'0_author_affs_0'] = QJE_scopus.iloc[i]['affiliations'].title().strip()
        print("False")
    else:

        # Split on new lines
        authors = QJE_scopus.iloc[i]['authors']
        authors_by_line = re.split(r"\n" , authors)
            
        
        for auth_no in range(len(authors_by_line)):
            QJE_scopus.loc[i,str(auth_no)+'_author'] = authors_by_line[auth_no]
            print(QJE_scopus.loc[i,str(auth_no)+'_author'])
            count = 0

        # For a-- b-- c--

            for association in association_list:
                # Checking for a-- or b-- in the authors line
                ratio = fuzz.partial_ratio(QJE_scopus.loc[i,str(auth_no)+'_author'] , association)
                if ratio > 95:
                    print("True")

                    #Splitting the affiliaitons by line
                    affiliations = QJE_scopus.iloc[i]['affiliations']
                    affiliations_by_line = re.split(r"\n" , affiliations)

                    # Looping through the affiliations by line checking for associated author a-- b--
                    for aff_no in range(len(affiliations_by_line)):
                    #Checking for a-- or b--
                        ratio_aff = fuzz.partial_ratio(affiliations_by_line[aff_no], association)
                        if ratio_aff > 95:
                            affil_auth = re.sub(r'a--|b--|c--|d--|e--|f--|g--|h--','',affiliations_by_line[aff_no]).title().strip()
                            QJE_scopus.loc[i,str(auth_no)+'_author_affs_'+str(count)] = affil_auth
                            print("True 95")         
                            count = count +1
                            print('count ' , count)
                        #else:
                            #print("False")

        #QJE_scopus.loc[i,str(auth_no)+'_author']

#QJE_scopus.iloc[[2846]].head()
```
